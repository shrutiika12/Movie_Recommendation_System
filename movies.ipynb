```python

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):
    # lowercase
    text = str(text).lower()
    
    # remove punctuataions
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # remove stopwords
    words = text.split();
    words = [word for word in words if word not in stop_words ]

    # lemmatize - words to the base form
    words =[lemmatizer.lemmatize(word) for word in words ]
    # list to string - join
    return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess)
```

```python
df['tags'][1]
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index = df['title'])
```

```python
# for cosine similarity
indices
```

```python
# Text vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features = 50000,ngram_range=(1,2) ,stop_words = 'english')
```

```python
 tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix 
```

```python
# Cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommendation(title, n =10):
    if title not in indices:
        return['Movie not found']

    idx = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx]
        
```

```python
recommendation('Toy Story')
```

```python
recommendation('The Shutter Island')
```
